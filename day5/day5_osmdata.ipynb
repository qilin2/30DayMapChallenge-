```r
#Data Imports
library(tidyverse)
library(ggplot2)
library(osmdata)
library(ggmap)
library(sf)
library(sysfonts)
library(showtext)
library(cowplot)
library(png)
```

```r
font_add_google("Patrick Hand SC")
font1 <- "Patrick Hand SC"
showtext_auto()
```

```r
download.file("https://i.imgur.com/Tpon3ny.png", "flag.png", mode = 'wb')
flag <- "flag.png"
```

```r
available_tags("shop")
```

```r
query <- getbb("Hong Kong") %>%
opq() %>%
add_osm_feature("shop", "convenience") %>%
add_osm_feature("name", "7-Eleven")

str(query)
```

```r
#Obtain resulting spatial data
shop <- osmdata_sf(query)

shop
```

```r
#Data Visualization
#Background Map
hk_map <- get_map(getbb("Hong Kong"), maptype = 'toner-background')

#Final Map
hk_711 <-
geom_sf(data = hk_rnatural)
geom_sf(data = shop$osm_points,
       inherit.aes = FALSE,
       color = 'red',
       fill = 'pink',
       alpha = 0.5,
       size = 4,
       shape = 15) +
theme_void()
hk_711
```

```r
#Building Query No.2
query2 <- getbb("Hong Kong") %>%
opq(timeout = 25*100) %>%
add_osm_feature("name", "7-Eleven") %>%
add_osm_feature("shop", "convenience")

#Query No.2
hk <- osmdata_sf(query2)

#Transparent Map with Dots!
hk_convenience <-
ggplot(hk$osm_points) +
geom_sf(color = '#ba0051',
       fill = '#d48c8c',
       alpha = 0.8,
       size = 2,
       shape = 21) +
theme_void() +
theme(plot.title = element_text(hjust = 0.5),
     plot.background = element_rect(fill = 'black', color = NA))
```

```r
hk_final <- ggdraw(hk_convenience) +
draw_label("7-Eleven Locations in Hong Kong", x = 0.5, fontfamily = font1, color = '#d48c8c', size = 27.5) +
draw_image(flag, x = 0.3, y = -0.0525, scale = .125) +
draw_label(label = "Creation: Brandon Qilin Hung | Data: OSMData", x = 0.5, y = 0.225, color = 'white', size = 12.5, fontfamily = font1)


hk_final
```
